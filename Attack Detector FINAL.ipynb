```python
data.groupby('type')['type'].count()
```

```python
sampler=np.random.permutation(len(data))
data=data.take(sampler)
data.head()
```

```python
labels_full=pd.get_dummies(data['type'], prefix='type')
labels_full
```

```python
data=data.drop(columns='type')
data.head()
```

```python
def standardize(df,col):
    df[col]= (df[col]-df[col].mean())/df[col].std()

data_st=data.copy()
for i in (data_st.iloc[:,:-1].columns):
    standardize (data_st,i)

data_st.head()
```

```python
train_data_st=data_st.values
train_data_st
```

```python
labels=labels_full.values
labels
```

# LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(train_data_st, labels, test_size=0.2, random_state=42)


lstm_model = Sequential()
lstm_model.add(LSTM(32, input_shape=(x_train.shape[1], 1), return_sequences=True))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(64, return_sequences=True))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(32))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(labels.shape[1], activation='softmax'))


lstm_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


lstm_model.summary()


```

```python

model_checkpoint = ModelCheckpoint('lstm_model_checkpoint.h5', monitor='val_loss', save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)


history = lstm_model.fit(x_train, y_train, validation_data=(x_test, y_test),
                         epochs=10, batch_size=32, callbacks=[early_stopping, model_checkpoint])


lstm_model.save('lstm_model.h5')

lstm_model.save('G:\SIDDHESH\College Data\BE Sem1\Project\lstm_model_final.h5')
```

```python
lstm_model.load_weights('lstm_model_checkpoint.h5')
```

```python
test_loss, test_accuracy = lstm_model.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")
```

# PNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint

x_train, x_test, y_train, y_test = train_test_split(
    train_data_st, labels, test_size=0.25, random_state=42)


def create_progressive_nn(input_size, output_size):
    model = Sequential()
    model.add(Dense(64, input_dim=input_size, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(output_size, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

```

```python

progressive_nn = create_progressive_nn(train_data_st.shape[1], labels.shape[1])

model_checkpoint = ModelCheckpoint('Pnn_model_checkpoint.h5', 
                                   monitor='val_loss', 
                                   save_best_only=True)



```

```python
progressive_nn.fit(x_train, y_train, validation_data=(x_test, y_test),
                   verbose=2, epochs=10 , callbacks=[model_checkpoint])

progressive_nn.save('Pnn_model_final.h5')
```

```python
progressive_nn.load_weights('Pnn_model_checkpoint.h5')
```

```python
test_loss, test_accuracy = progressive_nn.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")
```

# DNN 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(train_data_st, labels, test_size=0.2, random_state=42)

model = Sequential()


model.add(Dense(128, input_dim=train_data_st.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))


model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))


model.add(Dense(labels.shape[1], activation='softmax'))


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


model.summary()



```

```python


model_checkpoint = ModelCheckpoint('dnn_model_checkpoint.h5', monitor='val_loss', save_best_only=True)


history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32, callbacks=[model_checkpoint])
                    
```

```python
model.load_weights('dnn_model_checkpoint.h5')
```

```python

test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained hybrid model
hybrid_model = load_model("lstm_model_checkpoint.h5")
```

```python
def preprocess_user_file(file_path):
    # Load the file
    user_data = pd.read_csv(file_path)
    
    # Standardize numerical columns
    for col in user_data.columns[:-1]:  # Exclude the last column (which contains labels)
        user_data[col] = (user_data[col] - user_data[col].mean()) / user_data[col].std()
    
    # Convert to numpy array
    user_data_array = user_data.values
    
    # Reshape for model input
    user_data_reshaped = user_data_array.reshape(user_data_array.shape[0], user_data_array.shape[1], 1)
    
    return user_data_reshaped, None  # Return None for additional features (if any)
```

```python
def predict_sequential_data(data):
    # Make predictions using the hybrid model
    predictions = hybrid_model.predict(data)
    return predictions
```

```python
def predict_user_file(file_path):
    user_data_reshaped, _ = preprocess_user_file(file_path)
    
    # Predict using the sequential data
    predictions = predict_sequential_data(user_data_reshaped)
    
    # Determine the predicted class
    predicted_class = predict_class(predictions)
    
    return predicted_class
```

```python

def predict_class(predictions):
    # Define thresholds for classifying the data
    attack_threshold = 0.5  # Example: if the probability of an attack exceeds 0.5, classify as an attack
    
    # Check if the highest probability exceeds the threshold
    if np.max(predictions) > attack_threshold:
        return "Attack"
    else:
        return "Normal"
```
